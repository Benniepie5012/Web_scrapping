```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=shoes&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_2_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_2_0_na_na_na&as-pos=2&as-type=TRENDING&suggestionId=shoes&requestId=5ec4b946-5e52-4dfe-8ae1-637b51683b86"
```

```python
r = requests.get(url)
```

```python
r.status_code
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
names = soup.find_all("a", class_ ="IRpwTa")
print(names)
```

```python
p_name = []
for i in names:
    name = i.text
    p_name.append(name)
print(p_name)
```

```python
prices = soup.find_all("div", class_ ="_30jeq3")
print(prices)
```

```python
price_list = []
for i in prices:
    price = i.text
    price_list.append(price)
print(price_list)
```

```python
#<div class="_3Ay6Sb"><span>38% off</span></div>
```

```python
percents = soup.find_all("div", class_ ="_3Ay6Sb")
print(percents)
```

```python
percent_list = []
for i in percents:
    percent = i.text
    percent_list.append(percent)
print(percent_list)
```

```python
print(len(p_name), len(price_list), len(percent_list))
```
