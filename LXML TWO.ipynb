```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://prsindia.org/covid-19/cases"
```

```python
r = requests.get(url)
```

```python
print(r)
```

```python
r.status_code
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
names = soup.find_all("th")
print(names)
```

```python
title_name = []
for i in names:
    title = i.text
    title_name.append(title)
print(title_name)
```

```python
df = pd.DataFrame(columns=title_name)
print(df)
```

```python
df
```

```python
rows = soup.find_all("tr")
print(rows)
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    row = [tr.text for tr in data]
    print(row)
    l = len(df)
    df.loc[l] = row
print(df)
   
```
